```python
import math
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA

# Function to download stock data using yfinance
def get_yf_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data

# Function to calculate financial indicators (if needed later)
def calculate_financial_indicators(stock_data):
    financial_indicators = {}
    financial_indicators['Daily_Return'] = stock_data['Adj Close'].pct_change()
    financial_indicators['Cumulative_Return'] = (1 + financial_indicators['Daily_Return']).cumprod() - 1
    financial_indicators['50Day_MA'] = stock_data['Adj Close'].rolling(window=50).mean()
    financial_indicators['200Day_MA'] = stock_data['Adj Close'].rolling(window=200).mean()
    return pd.DataFrame(financial_indicators)

# Function to make series stationary
def make_stationary(series):
    # First order differencing to make series stationary
    stationary = series.diff().dropna()
    return stationary

# Function to check for stationarity using ADF test
def check_stationarity(series):
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    # Determine if the series is stationary
    is_stationary = result[1] < 0.05

    return is_stationary

# Function to implement ARIMA
def apply_arima(data):
    model = ARIMA(data, order=(1,1,0))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=63)  # Forecast the next quarter
    stderr = math.sqrt(model_fit.sse / len(data))  # Approximate standard error
    return forecast, stderr

# Main execution block
if __name__ == "__main__":
    ticker = 'BZ=F'
    start_date = '2020-03-01'
    end_date = '2023-11-14'  # Your original end date

    # Fetching stock data
    stock_data = get_yf_data(ticker=ticker, start_date=start_date, end_date=end_date)

    # Make series stationary
    stationary_data = make_stationary(stock_data['Adj Close'])

    # Check if data is stationary
    if check_stationarity(stationary_data):
        print("The series is stationary.")
        forecast, stderr = apply_arima(stationary_data)

        # Creating date index for forecast data, starting from the next day
        next_day = stock_data.index[-1] + pd.Timedelta(days=1)
        forecast_index = pd.date_range(start=next_day, periods=len(forecast), freq='B')

        # Creating dataframe for forecast
        df_forecast = pd.DataFrame(forecast, index=forecast_index, columns=['Forecast'])

        # Visualize the forecast
        plt.figure(figsize=(12, 8))
        plt.plot(stock_data['Adj Close'], label='Historic Data')
        plt.plot(df_forecast['Forecast'], label='Forecasted Data')
        plt.title(f'ARIMA Model Forecast for {ticker} for Next Quarter')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("The series is not stationary. Additional differencing or other transformations may be required.")
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error

# Function to download and preprocess the stock data
def get_yf_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    # Ensure no zero or negative prices which cause problems in logs
    stock_data = stock_data[stock_data['Adj Close'] > 0]
    # Log return transformation
    stock_data['LogRet'] = np.log(stock_data['Adj Close']).diff()
    return stock_data.dropna()

# Function to check for stationarity using ADF test
def check_stationarity(series):
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    for key, value in result[4].items():
        print(f'Critical Value {key}: {value}')
    return result[1] < 0.05

# Function to evaluate an ARIMA model for a given order (p,d,q)
def evaluate_arima_model(series, arima_order):
    split_size = int(len(series) * 0.8)
    train, test = series[0:split_size], series[split_size:]
    history = [x for x in train]
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit()
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(test.iloc[t])
    error = mean_squared_error(test, predictions)
    return error

# Function to perform a grid search for p,d,q ARIMA model parameters
def optimize_arima(series, p_values, d_values, q_values):
    best_score, best_cfg = float('inf'), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = evaluate_arima_model(series, order)
                    if mse < best_score:
                        best_score, best_cfg = mse, order
                    print('ARIMA%s MSE=%.5f' % (order, mse))
                except:
                    continue
    print('Best ARIMA%s MSE=%.5f' % (best_cfg, best_score))
    return best_cfg

# Main code
if __name__ == "__main__":
    # Set the ticker, start date, and end date
    ticker = 'BZ=F'
    start_date = '2020-01-01'
    end_date = '2023-11-15'

    # Get the data
    data = get_yf_data(ticker, start_date, end_date)

    # Check if the series is stationary
    if check_stationarity(data['LogRet']):
        print('Series is stationary.')
        # Grid search hyperparameters for the ARIMA model
        p_values = [0, 1, 2, 4, 6, 8, 10]
        d_values = range(0, 3)
        q_values = range(0, 3)
        best_order = optimize_arima(data['LogRet'], p_values, d_values, q_values)

        # Fit the ARIMA model
        model = ARIMA(data['LogRet'], order=best_order)
        model_fit = model.fit(disp=0)

        # Forecast the next 5 values
        forecast_steps = 5
        forecast = model_fit.forecast(steps=forecast_steps)[0]
        forecast_index = pd.date_range(start=data.index[-1], periods=forecast_steps + 1, closed='right')
        forecast_series = pd.Series(np.exp(forecast).cumprod() * data['Adj Close'].iloc[-1], index=forecast_index)

        # Plot the actual and forecasted prices
        plt.figure(figsize=(12,6))
        plt.plot(data['Adj Close'], label='Historical')
        plt.plot(forecast_series, color='red', label='Forecast')
        plt.title('Stock Price Forecast')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.legend()
        plt.show()
    else:
        print('Series is not stationary.')
```
