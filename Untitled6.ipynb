```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('AMD.csv')

print(df.describe())
print(df.isnull().sum())

df['Year'] = df['Date'].dt.year
Year = df['Year'].unique()
# Calculate average 'Adj Close' price per year
yearly_averages = df.groupby('Year')['Adj Close'].mean()

# Create a bar chart
yearly_averages.plot(kind='bar', figsize=(10, 6), color='royalblue', alpha=0.7)

# Customize the chart
plt.title('Average Adjusted Close Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Adjusted Close Price')
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf

# Function to download stock data using yfinance
def get_yf_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data

# Function to calculate financial indicators
def calculate_financial_indicators(stock_data):
    financial_indicators = {}
    financial_indicators['Daily_Return'] = stock_data['Adj Close'].pct_change()
    financial_indicators['Cumulative_Return'] = (1 + financial_indicators['Daily_Return']).cumprod() - 1
    financial_indicators['50Day_MA'] = stock_data['Adj Close'].rolling(window=50).mean()
    financial_indicators['200Day_MA'] = stock_data['Adj Close'].rolling(window=200).mean()
    return pd.DataFrame(financial_indicators)

# Function to simulate GBM
def simulate_gbm(S0, mu, sigma, T, dt, num_paths):
    num_steps = int(T/dt)
    paths = np.zeros((num_steps + 1, num_paths))
    paths[0] = S0

    # Loop to incrementally build paths
    for i in range(1, num_steps + 1):
        dW = np.random.normal(0, np.sqrt(dt), num_paths)
        paths[i] = paths[i - 1] * np.exp((mu - 0.5 * sigma ** 2) * dt + sigma * dW)

    return paths

if __name__ == "__main__":
    ticker = 'BZ=F'
    start_date = '2020-03-01'
    end_date = '2023-08-03'

    # Fetching stock data
    stock_data = get_yf_data(ticker=ticker, start_date=start_date, end_date=end_date)
    financial_indicators = calculate_financial_indicators(stock_data) # Calculating financial indicators from fetched stock data
    # Calculating rolling mean & std dev for financial indicators
    rolling_mean = stock_data['Adj Close'].rolling(window = 22).mean()
    rolling_std = stock_data['Adj Close'].rolling(window = 22).std()
    # Plotting the moving average technical analysis
    plt.plot(stock_data['Adj Close'], color = 'blue', label = 'Original')
    plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean')
    plt.plot(rolling_std, color = 'black', label = 'Rolling Std')
    plt.legend(loc = 'best')
    plt.title('Rolling Mean & Rolling Standard Deviation')
    plt.show()
    # Stochastic Simulation Parameters
    S0 = stock_data['Adj Close'][-1]  # Initial stock price at the last available date
    mu = financial_indicators['Daily_Return'].mean()  # Average daily return
    sigma = financial_indicators['Daily_Return'].std()  # Daily return volatility
    T = 252 / 252  # Time horizon in years (252 trading days)
    dt = 1 / 252  # Time increment (one trading day)
    num_paths = 10  # Number of simulation paths to generate

    # Stochastic Simulation
    paths = simulate_gbm(S0, mu, sigma, T, dt, num_paths)

    # Make a new time index for the forecast period
    forecast_period = pd.date_range(start=stock_data.index[-1], periods=253, freq='B')

    # Plot the simulation results
    plt.figure(figsize=(10, 6))
    for i in range(num_paths):
        plt.plot(forecast_period, paths[:, i])

    plt.title(f'GBM Simulation of {ticker} Stock Price for Next Quarter')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.grid(True)
    plt.legend()
    plt.show()
    # Plot the actual price history
    plt.plot(stock_data.index, stock_data['Adj Close'], label='Historic prices')

    # Plot the simulated paths
    for i in range(num_paths):
        plt.plot(forecast_period, paths[:, i])

    plt.title(f'GBM Simulation of {ticker} Stock Price')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.grid(True)
    plt.legend()

    plt.show()

    analysis_df = calculate_financial_indicators(stock_data) # Changing financial indicators to analysis_df to work with code given in class

    plt.figure(figsize=(12,6))
    # Plot the stock price and moving averages
    plt.subplot(2, 1, 1)
    plt.plot(stock_data['Adj Close'], label=ticker)
    plt.plot(analysis_df['50Day_MA'], label='50-Day MA', linestyle='dashed')
    plt.plot(analysis_df['200Day_MA'], label='200-Day MA', linestyle='dashed')
    plt.title(f'{ticker} Stock Price and Moving Averages')
    plt.legend()

    # Cumulative Returns
    plt.subplot(2, 1, 2)
    plt.plot(analysis_df['Cumulative_Return'])
    plt.title(f'{ticker} Cumulative Returns')
    plt.xlabel('Date')

    plt.tight_layout()
    plt.show()

    # Plot the autocorrelation of the adjusted close price
    fig, ax = plt.subplots(figsize=(10, 5))
    plot_acf(stock_data['Adj Close'].dropna(), ax=ax, lags=40)
    plt.show()


```
